```
from flask_ngrok import run_with_ngrok
from flask import Flask, request
app = Flask(__name__)
from flask_cors import CORS
run_with_ngrok(app)
CORS(app)
from tensorflow.keras.models import load_model
#from keras.models import load_model
import cv2
import base64
import numpy as np 
import json
from flask import jsonify

def funtest(image_data):
  model_new = load_model('/content/drive/My Drive/Colab Notebooks/model_fake_original_detectionimages_50epoch.h5')

  imgdata = base64.b64decode(image_data)
  filename = 'temp.jpg'  # I assume you have a way of picking unique filenames
  with open(filename, 'wb') as f:
    f.write(imgdata)
  #img = cv2.imread('/content/drive/My Drive/Colab Notebooks/currency_images/train2000/original_2000/output/fake50.jpg')
  img = cv2.imread("temp.jpg")
  img = cv2.resize(img,(300,150))
  img = np.reshape(img,[1,300,150,3])
  classes = model_new.predict(img)
  for item in classes[:10]:
    if (item[0]>.799):
      return ("Fake")
    else:
      return ("Real")
@app.route("/test", methods=['GET', 'POST'])
def hello_test():
    return jsonify ({'msg':"hello"})


@app.route("/", methods=['GET', 'POST'])
def hello():
    print ("Request is recieved")
    content = request.get_json()
    img_data = content["image"]
    #print (img_data)
    print("Processing the data")
    classification_result = funtest(img_data)
    print("Data Processed")
    print ("Output is", classification_result)
    dict_r = {"msg": classification_result}
    return jsonify(dict_r)
app.run()
```

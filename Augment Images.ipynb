```
import Augmentor

path_to_data = "/content/drive/My Drive/Colab Notebooks/currency_images/train2000/original_2000"

# Create a pipeline
p = Augmentor.Pipeline(path_to_data)
```

```
p.flip_left_right(probability=0.4)
p.rotate(probability=.5, max_left_rotation=5, max_right_rotation=10)
# Now we add a vertical flip operation to the pipeline:
p.flip_top_bottom(probability=0.8)
# Add a rotate90 operation to the pipeline:
p.rotate90(probability=0.5)
p.skew_tilt(probability=0.5)
p.skew_left_right(probability=0.5)
p.skew_top_bottom(probability=0.5)
p.skew_corner(probability=0.5)
p.skew(probability=0.5)
p.random_distortion(probability=0.5,grid_width=50,grid_height=50,magnitude=32)
p.rotate_random_90(probability=0.5)
p.flip_random(probability=0.5)
p.sample(5000)
p.process()
```
